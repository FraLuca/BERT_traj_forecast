## Focalizziamo su un singolo utente

```python
sub_df = df_edeka[df_edeka['tag_id'] == '0x00205EFB12AC']
sub_df
```

```python
sub_df = sub_df.sort_values(by='time')
sub_df
```

```python
def cast_time(x):
    return datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S')
```

```python
sub_df['datetime'] = sub_df['time'].apply(cast_time)
```

```python
sub_df.head()
```

```python
def compute_delta_time(x):
    date_1 = sub_df.iloc[0]['datetime']
    date_2 = x
    time_delta = (date_2 - date_1)
    return time_delta.total_seconds()
```

```python
sub_df['deltatime'] = sub_df['datetime'].apply(compute_delta_time)
```

```python
def convert_x_to_visualize(x):
    a = df_edeka['x'].min()
    b = df_edeka['x'].max()
    num_pixel_on_x = binary_matrix.shape[0]
    return num_pixel_on_x*(x-a)/(b-a)

def convert_y_to_visualize(y):
    a = df_edeka['y'].min()
    b = df_edeka['y'].max()
    num_pixel_on_y = binary_matrix.shape[1]
    return num_pixel_on_y*(1-(y-a)/(b-a))
```

## Filter just the useful sequences

```python
df_edeka['tag_id'].unique()
```

```python
sub_df = df_edeka[df_edeka['tag_id'] == '0x00205F15109C']
sub_df = sub_df.sort_values(by='time')
sub_df['datetime'] = sub_df['time'].apply(cast_time)
sub_df['deltatime'] = sub_df['datetime'].apply(compute_delta_time)
```

```python
threshold = 120

list_of_seq = []
curr_seq = []

for row in range(len(sub_df)-1):
    if sub_df.iloc[row+1]['deltatime'] - sub_df.iloc[row]['deltatime'] <= threshold:
        if len(curr_seq) == 0:
            curr_seq.append(sub_df.iloc[row])
            curr_seq.append(sub_df.iloc[row+1])
        else:
            curr_seq.append(sub_df.iloc[row+1])
    else:
        if len(curr_seq) != 0:
            list_of_seq.append(curr_seq)
            curr_seq = []
```

```python
list_of_time = []

for i in range(len(list_of_seq)):
    out_df = pd.DataFrame(list_of_seq[i]).drop_duplicates(subset=['datetime'])
    list_of_time.append(out_df.iloc[-1]['deltatime'] - out_df.iloc[0]['deltatime'])
```

```python
min_time_length = 10

new_seq_df = pd.DataFrame(columns = ['tag_id', 'time', 'x', 'y', 'description', 'datetime', 'deltatime', 'seq_idx'])

seq_idx = 0

for i in range(len(list_of_time)):
    if list_of_time[i] >= min_time_length:
        out_df = pd.DataFrame(list_of_seq[i]).drop_duplicates(subset=['datetime'])
        out_df['seq_idx'] = [seq_idx]*len(out_df)
        seq_idx += 1
        new_seq_df = pd.concat([new_seq_df, out_df], ignore_index=True, sort=False)
```

```python
from PIL import Image, ImageDraw
```

```python
def fig2data( fig ):
    """
    @brief Convert a Matplotlib figure to a 4D numpy array with RGBA channels and return it
    @param fig a matplotlib figure
    @return a numpy 3D array of RGBA values
    """
    # draw the renderer
    fig.canvas.draw( )

    # Get the RGBA buffer from the figure
    w,h = fig.canvas.get_width_height()
    buf = np.frombuffer( fig.canvas.tostring_argb(), dtype=np.uint8 )
    buf.shape = ( w, h,4 )

    # canvas.tostring_argb give pixmap in ARGB mode. Roll the ALPHA channel to have it in RGBA mode
    buf = np.roll( buf, 3, axis = 2 )
    return buf

 
def fig2img( fig ):
    """
    @brief Convert a Matplotlib figure to a PIL Image in RGBA format and return it
    @param fig a matplotlib figure
    @return a Python Imaging Library ( PIL ) image
    """
    # put the figure pixmap into a numpy array
    buf = fig2data( fig )
    w, h, d = buf.shape
    return Image.frombytes( "RGBA", ( w ,h ), buf.tostring( ) )
```

```python
def generate_gif(seq_idx, name_file, speed=5, loop=0):
    
    images = []
    duration = (new_seq_df[new_seq_df['seq_idx']==seq_idx].iloc[-1]['deltatime'] - new_seq_df[new_seq_df['seq_idx']==seq_idx].iloc[0]['deltatime'])/speed

    xy_df = new_seq_df[new_seq_df['seq_idx']==seq_idx][['x', 'y']]

    rescaled_x = xy_df['x'].apply(convert_x_to_visualize)
    rescaled_y = xy_df['y'].apply(convert_y_to_visualize)

    for idx in range(len(rescaled_x)):

        fig, ax = plt.subplots(figsize = (10, 7))
        plt.imshow(binary_matrix)

        plt.scatter(rescaled_x[:idx], rescaled_y[:idx])
        plt.plot(rescaled_x[:idx], rescaled_y[:idx], '--')

        images.append(fig2img(fig))

    images[0].save(name_file+'.gif', save_all=True, append_images=images, loop=loop, duration=duration, optimize=True)
```

```python
generate_gif(seq_idx=3, name_file='seq_3', speed=5, loop=0)
```
